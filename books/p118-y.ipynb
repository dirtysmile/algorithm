```python
map = [[1,1,1,1],[1,0,0,1],[1,1,0,1],[1,1,1,1]]
look = [0,1,2,3]
man = [1,1,0]

dx = [1,1,0,-1]
dy = [-1,0,1,0]

visit = [[0]*4 for _ in range(4)]
direction = man[2]
turn_time = 0
move_cnt = 1

def turn_left():
    global direction
    global turn_time

    turn_time += 1
  
    direction =  direction-1
    if direction == -1:
        direction = 3
    
    

def move_man():
    global man
    global direction
    global visit
    global turn_time
    global move_cnt
    
    x = man[0]
    y = man[1]
    
    visit[y][x] = 1
    
    print("---------visit ---------")
    for v in visit :
        print(v)
    
    turn_left()
    print("direction : " + str(direction))
    x = x+dx[direction]
    y = y+dy[direction]
    
    print('turn_time : ' + str(turn_time))
    print(x,y)
#     print("map : " + str(map[x][y]))
#     print(map[2][1])
    
    print(map[y][x])
    if(visit[y][x]==0 and map[y][x]==0):
#       이동할 수 있는 구간.
        print('육지다')
        man[0] = x
        man[1] = y
        turn_time = 0
        move_cnt += 1

    else:
        print('바다')
        
    if(turn_time == 4):
        return False;
    

    return True
    

while True:    
    if(not move_man()):
        print("move_cnt : " + str(move_cnt))
        break;

```
